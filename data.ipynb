```python
import pandas as pd
import numpy as np
class loader:
    def __init__(self,path):
        self.path =path
       
class Csv_loader(loader):
    def load_data(self):
        return pd.read_csv(self.path)
```

```python
class cleanner(Csv_loader):
    def drop_na(self,subset =None):
        return self.load_data().dropna(subset = subset) if subset else self.load_data().dropna()
    def fillna(self,**kwargs):
        self.kwargs = kwargs
        return self.load_data().fillna(kwargs)
    def normalize_text(self, column):
        df = self.load_data()
        df[column] = df[column].str.lower().str.strip()
        return df
    def cast_columns_type(self, **kwargs):
      return self.load_data().astype(kwargs) 

    def rename_col(self,**kwargs):
        self.kwargs = kwargs
        return self.load_data().rename(kwargs)
    
    def remove_outliers_iqr(self, column):
        df = self.load_data()
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
        return filtered_df


    

```

```python
import pandas as pd
import numpy as np

def clean_installs(value):
    try:
        if isinstance(value, str):
            value = value.replace(',', '').replace('+', '').strip()
            return int(value)
        elif isinstance(value, int):
            return value
        else:
            return np.nan
    except:
        return np.nan

outlier_free_df['Installs'] = outlier_free_df['Installs'].apply(clean_installs)

    
```

```python
def clean_size(value):
    try:
        if isinstance(value, str):
            value = value.strip().upper()
            if value == 'Varies with device' or value == '':
                return np.nan
            elif 'M' in value:
                return float(value.replace('M', '')) * 1_000_000
            elif 'K' in value:
                return float(value.replace('K', '')) * 1_000
            else:
                return float(value)
        elif isinstance(value, (int, float)):
            return float(value)
        else:
            return np.nan
    except:
        return np.nan

outlier_free_df['Size'] = outlier_free_df['Size'].apply(clean_size)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Define categorical and numeric columns
categorical_cols = ['Category', 'Type', 'Content Rating', 'Genres']
numeric_cols = ['Rating', 'Reviews', 'Size', 'Installs', 'Price']

# Preprocessing: scale numerics, one-hot categoricals
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=True), categorical_cols)
    ]
)



# Build pipeline with optimal number of clusters
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('cluster', KMeans(n_clusters=3, random_state=42))
])

# Fit the pipeline
pipeline.fit(outlier_free_df)

# Transform features for evaluation metrics
X_transformed = pipeline.named_steps['preprocessing'].transform(outlier_free_df)
labels = pipeline.named_steps['cluster'].labels_

# Silhouette Score
sil_score = silhouette_score(X_transformed, labels)
print("Silhouette Score:", sil_score)

# Calinski-Harabasz Index (convert to dense if needed)
if hasattr(X_transformed, "toarray"):
    X_dense = X_transformed.toarray()
else:
    X_dense = X_transformed
ch_score = calinski_harabasz_score(X_dense, labels)
print("Calinski-Harabasz Index:", ch_score)

# Add cluster labels to dataframe
outlier_free_df['Cluster'] = labels

# Mean stats per cluster
cluster_summary = outlier_free_df.groupby('Cluster')[numeric_cols].mean()
print(cluster_summary)


```

```python

from sklearn.metrics import davies_bouldin_score
db_score = davies_bouldin_score(X_dense, labels)
print("Davies-Bouldin Index:", db_score)
```

```python
nltk.download([ 'wordnet', 'punkt'])
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
wn = WordNetLemmatizer()
# Create an instance of the PorterStemmer
ps = PorterStemmer()
def clean_text(text, lemmatize=True):
    #  Lowercase
    text = text.lower()
    
    # Remove Punctuation
    text = re.sub(r'[^\w\s]', '', text) 
    
    # Split text into tokens (words)
    words = nltk.word_tokenize(text)
    
    #  Remove Stopwords
    words = [word for word in words if word not in stopwrds]
    
    #  Lemmatize (Optional)
    
    words = [wn.lemmatize(word) for word in words]
        
    # Join the processed words back into a single string
    return " ".join(words)
```

```python
import pandas as pd
from io import StringIO
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np
df = user_df.copy()


TEXT_FEATURE = 'Cleaned'
NUMERICAL_FEATURES = ['Sentiment_Polarity', 'Sentiment_Subjectivity']
ALL_FEATURES = [TEXT_FEATURE] + NUMERICAL_FEATURES

# Drop any rows where the text feature is missing, as TF-IDF can't handle NaN
df.dropna(subset=[TEXT_FEATURE], inplace=True)



preprocessor = ColumnTransformer(
    transformers=[
        ('text_vec', TfidfVectorizer(max_features=1000, stop_words='english'), TEXT_FEATURE),
        ('num_scale', StandardScaler(), NUMERICAL_FEATURES)
    ],
    
    remainder='passthrough'
)

K_CLUSTERS = 3
kmeans = KMeans(n_clusters=K_CLUSTERS, random_state=42, n_init=10)

#  Create the full ML Pipeline: Preprocessing + Clustering
model = Pipeline(steps=[('preprocessor', preprocessor), ('cluster', kmeans)])

# Fit the model and predict clusters
print(f"Fitting K-Means model with K={K_CLUSTERS}...")
model.fit(df[ALL_FEATURES])
df['New_KMeans_Cluster'] = model.predict(df[ALL_FEATURES])

# --- Analysis and Output ---
print("\n--- New Clustering Results ---")
print("New Cluster Distribution:")
print(df['New_KMeans_Cluster'].value_counts().sort_index())

print("\n--- Cluster Summary by Average Polarity ---")
cluster_summary = df.groupby('New_KMeans_Cluster')['Sentiment_Polarity'].agg(['count', 'mean', 'median']).sort_values('mean', ascending=False)
print(cluster_summary)

# Display a few sample reviews from each new cluster to understand the grouping
print("\n--- Sample Reviews per New Cluster ---")
for i in range(K_CLUSTERS):
    sample_reviews = df[df['New_KMeans_Cluster'] == i]['Translated_Review'].head(2).tolist()
    avg_polarity = cluster_summary.loc[i, 'mean']
    print(f"\nCluster {i} (Avg Polarity: {avg_polarity:.3f}):")
    for review in sample_reviews:
        print(f"  - {review[:60]}...")
```

```python
feature_names = model.named_steps['preprocessor'].named_transformers_['text_vec'].get_feature_names_out()

cluster_centers = model.named_steps['cluster'].cluster_centers_

tfidf_center_weights = cluster_centers[:, :len(feature_names)]

print("--- Top 10 Descriptive Terms for Each Cluster (Optimal K=3) ---")

for i in range(K_CLUSTERS):
    # Get the indices of the top 10 weights for the current cluster center
    top_10_indices = tfidf_center_weights[i].argsort()[-10:][::-1]

    # Use the indices to find the actual words
    top_10_words = [feature_names[j] for j in top_10_indices]

    # Manually determine the cluster's sentiment label based on the mean polarity
    #  use the existing cluster summary from your previous step for context:
    # Cluster 2  was highest polarity, Cluster 0 was lowest.
    if i == 2:
        sentiment_label = "HIGH POSITIVE (Enthusiastic)"
    elif i == 0:
        sentiment_label = "NEGATIVE (Issue Focused)"
    else: # i == 1
        sentiment_label = "LOW POSITIVE/NEUTRAL (Informational)"


    print(f"\nCluster {i} ({sentiment_label}):")
    print(f"Keywords: {', '.join(top_10_words)}")


```
